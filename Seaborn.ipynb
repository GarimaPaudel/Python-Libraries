## Dist Plot

Plots with Bar Graph and Kernel Density Estimate

```python
sns.distplot(df['time_spend_company'])
```

```python
sns.distplot(df['left'])
```

```python
sns.distplot(df['average_montly_hours'])
```

Histograms represnt the data distribution by forming bins along the range of the data and then drawing bars to show the number of observations that fall in each bin

```python
df.describe()
```

```python
bins = [2,3,4,5,6,7,8,9,10]
sns.distplot(df['time_spend_company'], bins = bins)
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins = bins , kde = False)
```

```python
sns.distplot(df['time_spend_company'], bins = bins , hist = False)
```

```python
sns.distplot(df['time_spend_company'], bins = bins , rug = True)
```

```python
sns.distplot(df['time_spend_company'], bins = bins , rug = True)
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins = bins, rug = True,
            hist_kws = {'color':'red', 'edgecolor':'blue', 'linewidth':3, 'alpha' : 0.7 },
             kde_kws = {'color':'orange', 'linewidth':3 })
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins = bins, rug = True,
            color = 'green')
plt.xticks(bins)
```

## Scatter Plot

```python
titanic_df = sns.load_dataset('titanic')
titanic_df.head()
```

```python
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df)
```

```python
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'alive')
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'alive', style = 'class')
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'alive', style = 'class', palette = 'inferno')
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'alive', style = 'class', palette = 'gist_rainbow', alpha = 0.5)
plt.title('Titanic Data Analysis')
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(x = 'age', y = 'fare', data = titanic_df, hue = 'alive', style = 'class')
sns.lineplot(x = 'age', y = 'fare', data = titanic_df, color = 'green')
plt.title('Titanic Data Analysis')
```

## Bar Plots

```python
titanic_df.head(1)
```

```python
sns.barplot(x = 'class', y = 'fare', data = titanic_df)
```

A bar plot represents an estimate of central tendency for a numeric variable with the height of each rectangle and provides some indication of the uncertainity around that estimate using error bars. 

```python
sns.barplot(x = 'class', y = 'fare', data = titanic_df, hue = 'sex')
```

```python
sns.barplot(x = 'class', y = 'fare', data = titanic_df, hue = 'sex', palette = 'inferno')
```

```python
# confidence interval
sns.barplot(x = 'class', y = 'fare', data = titanic_df, hue = 'sex',
           palette = 'inferno',
           errorbar = ('ci',100),
           errcolor = 'red',
           errwidth = 10)
```

```python
sns.barplot(x = 'class', y = 'fare', data = titanic_df,
            hue = 'sex',
           palette = 'inferno',
           errorbar = ('ci',100),
           errcolor = 'red',
           saturation = 0.9)
```
