```python
# Dropping rows with NaN values

df1 = df.dropna()     # default axis = 0
```

```python
# Dropping columns with NaN values

df2 = df.dropna(axis=1)  # axis = 0 for rows and 1 for columns
```

```python
df2.shape
```

```python
df.dropna(how = 'any') # if any row value is null then remove that row
```

```python
df.dropna(how = 'all')   # if all row values is null then remove that row
```

```python
df.dropna(inplace = True) # if we put inplace = true , it will remove the null values and replace the original data frame
```

### Filling the null values

```python
mydf = pd.read_csv('titanic.csv')
mydf.head()
```

```python
mydf.isnull().sum()
```

```python
mydf.fillna(0)   # fills all null values with 0, we can also give other values
```

```python
mydf.fillna({'Age':'none','Cabin':0})  # we can even fill particular column's null values like this
```

### Forward fill and backward fill

```python
mydf.fillna(method = 'ffill') # forward fill fills the null value with the previous/preceeding value
```

```python
mydf.fillna(method = 'ffill', axis=1) # null values is filled with the previous columns value 
```

```python
mydf.fillna(method = 'ffill', axis=0)   # bfill for backward fill
```

### Filling with average value

```python
mydf['Age'].fillna(value = mydf['Age'].mean())
```

## Merge() Function

```python
import pandas as pd
df1 = pd.DataFrame({'Roll no.':[1,2,3,4,5],'Physics':[34,65,90,89,72]})
df1
```

```python
df2 = pd.DataFrame({'Roll no.':[1,2,3,4,5],'Chemistry':[40,70,85,80,75]})
df2
```

```python
pd.merge(df1,df2, on = 'Roll no.')
```

```python
pd.merge(df2,df1, on = 'Roll no.')
```

```python
pd.merge(df1 , df2)   # on - by default intersecting column is taken
```

```python
df3 = pd.DataFrame({'Roll no.':[1,2,3,6,7],'Physics':[34,65,90,89,72]})
df4 = pd.DataFrame({'Roll no.':[1,2,3,8,9],'Chemistry':[40,70,85,80,75]})
```

```python
pd.merge(df3,df4)    # only the intersecting roll.nos 
```

```python
pd.merge(df3, df4, how = 'left')
```

```python
pd.merge(df3, df4, how = 'right')
```

```python
pd.merge(df3, df4, how = 'outer')
```

## append() function in Pandas

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5],'Maths':[45,78,46,90,66],'Physics':[33,67,12,90,44]})
df2 = pd.DataFrame({'Roll No.':[6,7,8,9,10],'Maths':[78,73,45,90,69],'Physics':[23,67,88,0,98]})
```

```python
print(df1)
print(df2)
```

```python
df1.append(df2)
```

```python
pd.concat([df1, df2])
```

```python
pd.concat([df1, df2], ignore_index = True)
```

```python
pd.concat([df1, df2], ignore_index = True, sort=True)
```

## Pivot Table

```python
df = pd.read_csv('titanic.csv')
df.head()
```

```python
pd.pivot_table(df, index = 'Survived')    # values should be numerical value, only numerical values columns 
```

```python
pd.pivot_table(df, index = 'Survived', aggfunc='sum')   # by default aggfun = mean
```

```python
pd.pivot_table(df, index = 'Survived', aggfunc='count')
```

```python
pd.pivot_table(df, index = 'Survived', aggfunc='max')
```

```python
pd.pivot_table(df, index = 'Survived', columns= 'Pclass')
```
